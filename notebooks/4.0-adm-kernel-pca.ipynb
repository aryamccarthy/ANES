```python
df = pd.read_csv("../data/processed/2012.csv", index_col=0)
```

```python
scaler = Pipeline([
        ('imp', Imputer(strategy='mean')),
        ('scl', StandardScaler())
    ])
X = scaler.fit_transform(df)
```
